Importing Python libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn import metrics
import numpy as np
```

Reading data from csv file and extracting into Pandas dataframe

```python
nba_dataset = pd.read_csv('/content/nba_playoffs.csv')
```

To count rows and columns in a dataframe

```python
nba_dataset.shape
```

```python
nba_dataset.info()
```

To check missing values

```python
nba_dataset.isnull().sum()
```

```python
nba_dataset.head()
```

```python
X = nba_dataset.drop(['player','position','teams','born','country','total_games','hall_of_fame'],axis=1)
print (X)
```

```python
Y = nba_dataset['total_games']  # Target variable
```

Splitting Data into test and train set

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, random_state=2)
```

```python
Correlation = X_train.corr()
Correlation.style.applymap(lambda x: 'background-color : blue' if x > 0.7 else '')
```

# Correlation

* field_goals, free_shots, three_points_goals have highest impact
* Multi-Colinearity is Present

# Model training on Linear Regression

```python
lin_reg_model = LinearRegression()
lin_reg_model.fit(X_train, Y_train)
```

Predict on Test data and calculate score

```python
from sklearn.metrics import r2_score
Y_pred = lin_reg_model.predict(X_test)
r2 = r2_score(Y_test, Y_pred)

print('R-Squared score for model Performance on Test : ', np.round(r2,2)*100)
```
